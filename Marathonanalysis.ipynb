```python
import pandas as pd
import seaborn as sns
```

```python
# GOAL IS TO ANALYSE DATA FOR USA --50KM OR 50MI IN 2020
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES")
```

```python
df.dtypes
```

```python
df[df['Event distance/length']=='50km'] # Allows us to see rows with 50km event distance/length
```

```python
df[df['Event distance/length']=='50mi'] # Allows us to see rows with 50mi event distance/length
```

```python
# Combining two dataframes of 50km and 50mi
# df[(df['Event distance/length']=='50mi') | (df['Event distance/length']== '50km')]
# Alternate way
df[df['Event distance/length'].isin(['50km', '50mi'])]
# returns a boolean series
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
# & operator works this way (df1) & (df2) merge these two within a parantheses of their own
```

```python
df[df['Event name'] =='Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split().str.get(5).str.split('(').str.get(1).str.split(')').str.get(0)
# We write event name twice because the existing data frame after finding everglades results in an entire dataframe so we need to specify to access
# event name again
```

```python
df[df['Event name'].str.split().str.get(5).str.split('(').str.get(1).str.split(')').str.get(0) =='USA']
```

```python
# combining all filters
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split().str.get(5).str.split('(').str.get(1).str.split(')').str.get(0) =='USA')]
```

```python
df2.head()
```

```python
df2.shape
```

```python
# removing (USA) in Event name to visualise it better 
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df.head(10)
```

```python
#cleaning up athlete age
# df2['athlete_age'] = 2020 - df2['Athlete year of birth'].astype(int)  this doesnt work bcs pandas by default knows how to handle missing values
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head(10)
```

```python
# removing 'h' from athlete performance
df2['Athlete performance']=df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
# axis 1 to drop columns
```

```python
df2.head(10)
```

```python
# now we clean null values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2[df2.duplicated() == True]
# empty table shows which means there is no duplicates in our dataframe
```

```python
# resetting index
df2.reset_index(drop = True)
```

```python
df2.dtypes
```

```python
# converting float to integer type
df2['athletes_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#renaming columns
df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id'
})
```

```python
# re ordering the column
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_age', 
           'athlete_id', 'athlete_gender', 'athlete_performance', 'athlete_average_speed', 'year']]
```

```python
# finding two races in 2020 -- Sarosta and Everglades
df3[df3['athlete_id'] == 222509]
```
